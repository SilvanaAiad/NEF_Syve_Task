### Task 1

```python
#grouping the data by pod type and knowing biggest CPU and Memory value for each of the pods
#getting maximum values
POD_Comparison = kubectl_top_data.groupby('NAME_OF_POD')[['CPU','MEM']].max()
POD_Comparison = POD_Comparison.reset_index()
POD_Comparison
```

```python
#Rename the columns

POD_Comparison.columns = ["NAME_OF_POD" , "MAX_CPU" , "MAX_MEM" ]
POD_Comparison
```

```python
# creating new file for the new result 

result1 = POD_Comparison.to_csv('/NEF_Syve_Task/result1.txt', sep = "\t" , index = False, mode = 'w')
```

### Task 2

```python
#Open input2 file 
Boundary_Values = pd.read_table("/NEF_Syve_Task/input2.txt", delimiter=r"\s+", header= None)

#Rename the columns
Boundary_Values.columns = ['NAME_OF_POD' , 'ACC_CPU' , 'ACC_MEM']
Boundary_Values
```

```python
#result1 table 
POD_Comparison
```

```python
#Collecting all data in one table
Below_OR_Beyond = POD_Comparison.copy()
Below_OR_Beyond['ACC_CPU'] = Boundary_Values['ACC_CPU']
Below_OR_Beyond['ACC_MEM'] = Boundary_Values['ACC_MEM']
Below_OR_Beyond
```

```python
#Is it below or beyond the acceptable value
Below_OR_Beyond ['Compare_CPU'] = np.where(Below_OR_Beyond['MAX_CPU'] <= Below_OR_Beyond['ACC_CPU'] , True, False)
Below_OR_Beyond ['Compare_MEM'] = np.where(Below_OR_Beyond['MAX_MEM'] <= Below_OR_Beyond['ACC_MEM'] , True, False)
Below_OR_Beyond
```

```python
#Two list of pods that performed below their boundaries in terms of CPU and Memory.
#CPU
CPU_Below = Below_OR_Beyond.loc[Below_OR_Beyond['Compare_CPU'] == True , ['NAME_OF_POD']]
POD_CPU_Below = CPU_Below ['NAME_OF_POD'].tolist()

#MEMORY
MEM_Below = Below_OR_Beyond.loc[Below_OR_Beyond['Compare_MEM'] == True , ['NAME_OF_POD']]
POD_MEM_Below = MEM_Below['NAME_OF_POD'].tolist()
```

```python
#Two list of pods that went beyond their acceptable limits in terms of CPU and Memory.
#CPU
CPU_Beyond = Below_OR_Beyond.loc[Below_OR_Beyond['Compare_CPU'] == False , ['NAME_OF_POD']]
POD_CPU_Beyond = CPU_Beyond['NAME_OF_POD'].tolist()

#MEMORY
MEM_Beyond = Below_OR_Beyond.loc[Below_OR_Beyond['Compare_MEM'] == False , ['NAME_OF_POD']]
POD_MEM_Beyond = MEM_Beyond ['NAME_OF_POD'].tolist()
```

```python
#open result2 file
result2 = open('/NEF_Syve_Task/result2.txt',  mode = 'w')
result2.write("List of pods NOT Exceeded CPU limit: ")
for item in POD_CPU_Below:
  result2.write("\n\t%s" % item)
result2.write("\nList of pods NOT Exceeded MRM limit: ")
for item in POD_MEM_Below:
  result2.write("\n\t%s" % item)
result2.write("\nList of pods Exceeded CPU limit: ")
for item in POD_CPU_Beyond:
  result2.write("\n\t%s" % item)
result2.write("\nList of pods Exceeded MEM limit: ")
for item in POD_MEM_Beyond:
  result2.write("\n\t%s" % item)
result2.close()
```
